```python
#Instalar las librerias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Cargar los datos
data = pd.read_csv("D:/UNAD/Practica Analisis de Datos/framingham.csv")
```

```python
#Mostrar los datos
data
```

```python
# Se realiza la selección de los datos /
data[['currentSmoker','heartRate']].head()
```

```python
#Generar grafica de puntos
data[['currentSmoker','heartRate']].plot.scatter(x='currentSmoker',y='heartRate')
```

```python
# Vamos a pintar una función logistica sobre estos datos

# pruebas de parametro
w = 0.09
b = -3.6
```

```python
# puntos de la recta
x = np.linspace(0,data['heartRate'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
# grafica de la recta
data.plot.scatter(x='currentSmoker',y='heartRate')
plt.plot(x, y, color='black')
plt.ylim(0,data['currentSmoker'].max()*1.1)
plt.scatter(x, y, color='#A9E253')
# plt.grid()
plt.xlabel('Ritmo Cardiaco')
plt.ylabel('Fumador Actual')
plt.show()
```
